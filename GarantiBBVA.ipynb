### Encoding

```python
categoricals = [x for x in test_users.columns if test_users[x].dtype == 'object']
numerics = [x for x in test_users.columns if test_users[x].dtype != 'object']
```

```python
for x in categoricals:
    train_category  = set(test_users[x].unique())
    test_category   = set(test_users[x].unique())
    most_category = set.intersection(train_category, test_category)
    
    train_users.loc[~train_users[x].isin(most_category), x] = 'other'
    test_users.loc[~test_users[x].isin(most_category), x] = 'other'
```

```python
data_all = pd.concat([train_users, test_users], axis=0)
```

```python
for x in categoricals:
    data_all[x] = data_all[x].factorize()[0]
```

```python
data_all[categoricals] = data_all[categoricals].astype('category')
data_all[numerics] = data_all[numerics].fillna(-1)

train_users = data_all.loc[train_users.index, train_users.columns]
test_users = data_all.loc[test_users.index, test_users.columns]
```

### Training

```python
from sklearn.model_selection import cross_val_score, GridSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
```

```python
X = train_users.drop(columns=["moved_after_2019"])
y = train_users["moved_after_2019"]
```

```python
rf = RandomForestClassifier()
cv  = StratifiedKFold(shuffle=True)
```

```python
rf_score = cross_val_score(rf, X, y, cv=cv, scoring='accuracy', n_jobs=-1)
print(f'Scores mean: {np.mean(rf_score):.4f}')
print(f'Scores std: {np.std(rf_score):.4f}')
```
